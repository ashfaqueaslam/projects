```python
import pandas as pd


def get_area(offset_dist, *offset):

    def first_method(offset_dist, *offset):
        count = 0
        for i in range(len(offset)):
            if i == 0 or i == len(offset) - 1:
                count += offset[i]

            elif i % 2 == 1:
                count += offset[i] * 4

            elif i % 2 == 0:
                count += offset[i] * 2

        return (count * 2 * offset_dist) / 3

        
    def alternate_method(offset_dist, *offset):
        count = 0
        for i in range(len(offset[:3])):
            if i == 0:
                count += offset[i] * 5
            elif i == 1:
                count += offset[i] * 8
            elif i == 2:
                count -= offset[i]

        count = (count * 2 * offset_dist) / 12

        count2 = 0
        for i in range(1, len(offset)):
            if i == 1 or i == len(offset) - 1:
                count2 += offset[i]

            elif i % 2 == 0:
                count2 += offset[i] * 4

            elif i % 2 == 1:
                count2 += offset[i] * 2

        count2 = (count2 * 2 * offset_dist) / 3
        return count2 + count


    if len(offset) % 2 == 1:
        return first_method(offset_dist, *offset)

    return alternate_method(offset_dist, *offset)

def half_method(offset_dist,*offset):
        count = 0
        for i in range(3):
            if i == 0:
                count += offset[i] * 5
            elif i == 1:
                count += offset[i] * 8
            elif i == 2:
                count -= offset[i]

        return (count * 2 * offset_dist) / 12
    
df_offset=pd.read_excel('offset.xlsx')
df_offset
```

```python
offset_dist=float(input('enter offset dist: '))

areas={}

for i in range(len(df_offset.columns)):
    areas[f'stn{i}']=[half_method(offset_dist,*df_offset.iloc[:3,i]),
                      get_area(offset_dist,*df_offset.iloc[:3,i])]
    for j in range(4,len(df_offset)+1):  
        areas[f'stn{i}'].append(get_area(offset_dist,*df_offset.iloc[:j,i]))
    
    
stn_names=[f'stn{i}' for i in df_offset.columns]
bonjean={}
for i in range(len(stn_names)):
        bonjean[stn_names[i]]=areas[f'stn{i}'] 

df=pd.DataFrame(bonjean)
for i in range(len(df.index)):
    for j in range(len(df.columns)):
        if df.iloc[i,j]<0.0:
            df.replace(df.iloc[i,j],0.0,inplace=True)

indexes =[f'0-{i}' for i in range(1,len(df_offset))]
old_index=[i for i in range(len(df_offset)-1)]
index_dict=dict(zip(old_index,indexes))

df.rename(index = index_dict, inplace = True)
df.to_excel('bonjean.xlsx',index=True)

df




    
    




        
```
